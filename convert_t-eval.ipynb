```python
import json
def write_json(Data, f_p):
    f = open(f_p, "w", encoding="utf-8")
    json.dump(Data, f, ensure_ascii=False, indent=2)
    f.close()

def read_json(f_p):
    f = open(f_p, "r", encoding="utf-8")
    Data = json.load(f)
    return Data

```

```python
from pydantic import BaseModel, Field
from typing import List, Optional
from faker import Faker
import random

fake = Faker()

class Property(BaseModel):
    id: int
    name: str
    address: str

class PropertyDetails(BaseModel):
    title: str
    propertyType: str
    personCapacity: int
    rating: float
    latitude: float
    longitude: float

class Availability(BaseModel):
    date: str
    available: bool

class Review(BaseModel):
    comment: str
    rating: float
    date: str

class CheckoutPrice(BaseModel):
    accommodationCost: float
```

```python
def generate_properties(place: str) -> List[Property]:
    return [Property(id=i, name=f"{place} House {i}", address=fake.address()) for i in range(1, 4)]

def generate_property_details(propertyId: int) -> PropertyDetails:
    return PropertyDetails(
        title=f"Title of Property {propertyId}",
        propertyType=random.choice(['Entire home/apt', 'Private room', 'Shared room']),
        personCapacity=random.randint(1, 6),
        rating=round(random.uniform(3.0, 5.0), 2),
        latitude=float(fake.latitude()),
        longitude=float(fake.longitude())
    )

def generate_availability(propertyId: int) -> List[Availability]:
    return [Availability(date=fake.date_this_decade().isoformat(), available=random.choice([True, False])) for _ in range(10)]

def generate_reviews(propertyId: int) -> List[Review]:
    return [Review(comment=fake.text(max_nb_chars=200), rating=round(random.uniform(1.0, 5.0), 2), date=fake.date_this_year().isoformat()) for _ in range(3)]

def generate_checkout_price(propertyId: int, checkIn: str, checkOut: str) -> CheckoutPrice:
    return CheckoutPrice(accommodationCost=round(random.uniform(50.0, 500.0), 2))
```

```python
def search_property_by_place(place: str) -> List[Property]:
    return generate_properties(place)

def get_property_details(propertyId: int) -> PropertyDetails:
    return generate_property_details(propertyId)

def check_availability(propertyId: int) -> List[Availability]:
    return generate_availability(propertyId)

def get_property_reviews(propertyId: int) -> List[Review]:
    return generate_reviews(propertyId)

def get_property_checkout_price(propertyId: int, checkIn: str, checkOut: str) -> CheckoutPrice:
    return generate_checkout_price(propertyId, checkIn, checkOut)
```
